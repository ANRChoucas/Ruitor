```python
def uriToName(uri, *args, **kwargs):
    uriDelimiter = kwargs.get("uriDelimiter", "#")

    swich = {
        rdflib.term.URIRef: lambda x: x.partition(uriDelimiter)[2],
        rdflib.term.BNode: lambda x: x.toPython(),
        rdflib.term.Literal: lambda x: x
    }

    return swich[type(uri)](uri)
```

```python
def definedClass(uri):
    classname = uriToName(uri)
    cls = _RDFClasses.get(classname, _MyRDFClasses.get(classname))
    return cls
```

```python
def create_rdfClass(name, values):
    
    attr_dct = {}
    
    for i, v in values.items():
        
        cls = definedClass(i)
        if cls:
            attr_dct[uriToName(i)] = cls(v)
        else :
            attr_dct[uriToName(i)] = type(uriToName(i), (), {"machin":v})
            
    return attr_dct
```
